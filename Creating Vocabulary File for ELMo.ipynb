## Reading File

```
from google.colab import drive
import numpy as np
import pandas as pd
drive.mount('/content/gdrive')
bengali_news_after_preprocessing = pd.read_pickle('/content/gdrive/My Drive/Projects/Bengali Text Classification/Bengali_Text_after_preprocessing.pkl')
```

## X contains the numpy array format of text

```
X = bengali_news_after_preprocessing.iloc[:,:].values
```

## Following line removes nonetype values from numpy array

```
X = X[X != np.array(None)]
```

## Following lines counts the tokens in data and sort them

```
unique, idx, cnt = np.unique(X, return_index = True, return_counts=True)
```

```
from operator import itemgetter
cnt = 1 / cnt
sorted_vocab = sorted(zip(unique, cnt, idx), key=itemgetter(1, 2))
```

```
unique, count, index = zip(*sorted_vocab)
```

```
len(unique)
```

## Finally vocabularies are saved in vocab.txt file

```
with open('/content/gdrive/My Drive/Projects/Bengali Text Classification/vocab.txt', 'w') as vocab:
  for i in unique:
    vocab.write(i+'\n')
```
