### Removement of stop words from texts.

```
STOP_WORDS = pd.read_json('/content/gdrive/My Drive/Projects/Bengali Text Classification/stopwords-bn.json',encoding='utf-8')
stops = [i for i in STOP_WORDS.values]
```

```
selected_news = bengali_news
content = [con['content'] for con in selected_news]
```

### Removement of regular expressions from texts and saving it into pickle file.

```
c = 0
x = 200
def sentence_to_wordlist(sentence, filters="!\"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n?,।!‍.'0123456789০১২৩৪৫৬৭৮৯‘\u200c–“”…‘"):
    translate_dict = dict((c, ' ') for c in filters)
    translate_map = str.maketrans(translate_dict)
    wordlist = sentence.translate(translate_map).split()
    global c,x;
    c+=1
    if c>x:
        print(x,end=' ')
        x+=100
    return list(filter(lambda x: x not in stops, wordlist))

#newsVec = [nltk.word_tokenize(text) for text in title]

filter_text = [sentence_to_wordlist(text) for text in content]
dataFrame = pd.DataFrame(filter_text)
dataFrame.to_pickle('/content/gdrive/My Drive/Projects/Bengali Text Classification/Bengali_Text_after_preprocessing.pkl')
```
